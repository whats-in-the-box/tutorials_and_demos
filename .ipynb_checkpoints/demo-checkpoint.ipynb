Import data and data wrangling.

## Check missing data. 

```r
d1 <- fh[-1,] # removing label from the dataset, just data left. 

sum(d1==0) ## Total missing values
sum(d1 ==0)/(dim(d1)[1]*dim(d1)[2])
```

The dataset contains 10448 missing values, missing data percentage is 17%, which is acceptable for raw peak intensity table. We want to convert all zeros to NA. 

```r
d1[d1 == 0] <- NA ## Change zeros to NA
sum(is.na(d1))
```

## Access feature presence/absence. 

As we stated, a feature is deemed present when missing values of a feature accounts for less than 50% of sample space, all sample or within each group. 

```r
d1$nacount <- as.matrix(apply(d1, 1, function(x) sum(is.na(x)))) #count the missing signal values by feature
d1$nacount_Native = as.matrix(apply(d1[, 1:6], 1, function(x) sum(is.na(x))))
d1$nacount_Semi = as.matrix(apply(d1[, 7:12], 1, function(x) sum(is.na(x))))

head(d1)
```

Scenario 1, keep a feature when it has na less than 50% of all samples.

```r
d2 <- d1[which(d1$nacount < round((length(d1)-3)*.5)), ] %>%
        select(-c(nacount, nacount_Semi, nacount_Native))
sum(is.na(d2)) # number of missing values
sum(is.na(d2))/(dim(d2)[1]*dim(d2)[2]) # percentage missing data
```

Scenario 2, keep a feature when it has na less than 50% of either group. 

```r
d2_2 = d1[which(d1$nacount_Native < round(6*.5) & d1$nacount_Semi < round(6*.5)), ] %>%
        select(-c(nacount, nacount_Semi, nacount_Native))
sum(is.na(d2_2)) # number of missing values
sum(is.na(d2_2))/(dim(d2_2)[1]*dim(d2_2)[2]) # percentage missing data
```

## Missing data imputation.

We are gonna impute both datasets, because their missing value percentages are both low enough. We will use k-nearest neighbors for imputation. This is how we decide k, square root of number of features. 

```r
# decide K (how many neighbor to use)
k <- d2_2 %>% length() %>% sqrt() # square root of number of features
if (k>=3){
        if (is.integer(k)) k = k else k = ceiling(k)
} else {k = 3}

k
```

Scenario 1:

```r
imputed_meta_d1 <- impute::impute.knn(as.matrix(d2), k = k, rowmax = 0.5,
                              colmax = 0.8, maxp = 1500)

imputed_d1 <- as.data.frame(imputed_meta_d1[["data"]], colnames = TRUE)
sum(imputed_d1 == 0)
sum(imputed_d1 < 0)
```

There is 8 zeros after imputation. This usually means there were too much missing values to begin with, and we are pushing the limit of KNN. 

Scenatio 2:

```r
imputed_meta_d1 <- impute::impute.knn(as.matrix(d2_2), k = k, rowmax = 0.5,
                              colmax = 0.8, maxp = 1500)

imputed_d1 <- as.data.frame(imputed_meta_d1[["data"]], colnames = TRUE)
sum(imputed_d1 == 0)
sum(imputed_d1 < 0)
```

There is no missing values after imputation for scenario 2, so this is the data we will continue with. Check the data distribution after imputation. 

```r
MASS::truehist(unlist(imputed_d1), main = "Imputed", nbins = 50) # check data distribution
```

This distribution is very typical for mass spec-generated data. Next step is data processing. 

## Log2 transformation. 

```r
log2_d1 <- log2(type.convert(imputed_d1))
MASS::truehist(unlist(log2_d1), main = "log2 transformed", nbins = 50) # check data distribution
car::qqPlot(unlist(log2_d1), main = "log2 transformed",xlab = "Theoretical Quantiles", ylab = "Sample Quantiles")
```

This already improves the data distribution quite a lot. But for the sake of demonstration, we will proceed with eigenMS normalization, which is implemented using package `ProteoMM`. Again, `ProteoMM` is a bioconductor package. 

## EigenMS normalization

```r
if (!requireNamespace("BiocManager", quietly = TRUE))
    install.packages("BiocManager")

if (!requireNamespace("ProteoMM", quietly = TRUE))
    BiocManager::install("ProteoMM")

suppressPackageStartupMessages(library("ProteoMM"))
```

```r
d4 <- cbind.data.frame(row.names(log2_d1), log2_d1) # check data structure, feature by sample
head(d4)
```

This following section of code is based on `ProteoMM` documentation. You can browse package vignette by typing this line of code in R console:
`browseVignettes("ProteoMM")`

```r
rowmeta <- d4[1,]
rowmeta2 <- as.data.frame(t(colnames(d4)))
colmeta <- d4[-1,1]

intsCols <- 2:dim(d4)[2]
m_logInts = ProteoMM::make_intencities(d4, intsCols)

metaCols <- 1
m_mets.info = as.matrix(ProteoMM::make_meta(d4, metaCols))

grps = as.factor(t(labels_d1))
m_nummiss = sum(is.na(m_logInts)) #check zeros

set.seed(42)
hs_m_ints_eig1 = eig_norm1(m=m_logInts,treatment = grps,
                           prot.info=m_mets.info)
hs_m_ints_norm_1bt = eig_norm2(rv=hs_m_ints_eig1)
```

The dataframe after eigenMS normalization is named `norm_d1`, and we will check its distribution as well. 

```r
norm_d1 <- as.data.frame(hs_m_ints_norm_1bt[["norm_m"]])
par(mfrow=c(1,1)) # after last step, the R graphic space are devided to 3 by 2 grid, this step is simply to restore. 
```

```r
MASS::truehist(unlist(t(norm_d1)), main = "EigenMS normalization", nbins = 50)
car::qqPlot(unlist(t(norm_d1)),main = "log2-EigenMS",xlab = "Theoretical Quantiles", ylab = "Sample Quantiles")
```

This data distribution looks normal based on histogram and qq plot, so we will stop here and proceed with univariate and multivariate analysis. 

Other normalization methods include rank invariant normalization implemented using `lumi`, cubic spline normalization implemented using `affy`. If normalization step cannot transform the data to a near Gaussian distribution, we will follow up with scaling step, implemented using `MetaboAnalyze`.

## Univariate analysis, t-test and Wilcoxon rank sum test. 

```r
d5 <- rbind.data.frame(labels_d1, norm_d1)

C <- d5[-1,which(d5[1,]=="Naive")]; P <- d5[-1,which(d5[1,]=="Semi_immue")]

wt <- data.frame(Doubles=double(),Doubles=double(),Doubles=double(),Doubles=double(),Doubles=double())

# Calculate p-values from both t-test and Wilcoson rank sum test. 
# P-values will need to be adjusted with Benjamini-Hochberg correction (i.e., FDR)
for (i in 1:dim(norm_d1)[1]){
        wt[i,1] <- t.test(type.convert(t(P[i,])),type.convert(t(C[i,])))$p.value
        wt[i,3] <- wilcox.test(type.convert(t(P[i,])),type.convert(t(C[i,])))$p.value
        az1 <- 2^(type.convert(C[i,])); az1L <- length(az1)
        az2 <- 2^(type.convert(P[i,])); az2L <- length(az2)
        wt[i,5] <- gtools::foldchange(sum(az2)/az2L, sum(az1)/az1L)
        #print(az1)
}
wt[,2] <- as.data.frame(p.adjust(wt[,1], method = 'BH'))#, n = length(p))
wt[,4] <- as.data.frame(p.adjust(wt[,3], method = 'BH'))#, n = length(p))

colnames(wt) <- c("pT", "BHT", "pW", "BHW", "FC(lin)")
row.names(wt) <- rownames(norm_d1)

# Calculate log2FoldChange
log2FoldChange =gtools::foldchange2logratio(wt$`FC(lin)`,base = 2)
wt = cbind(wt,log2FoldChange)
head(wt)
```

This is the statistical result table after univariate analysis, BH-corrected t-test and Wilcoxon rank sum tests. Next step is to screen for features that passed BHT < 0.05 & BHW < 0.05 criteria. 

```r
sum(wt[,1] < 0.05 & wt[,3] < 0.05)
sum(wt[,2] < 0.01 & wt[,4] < 0.01)
sig_wt <- wt[which(wt[,2] < 0.01 & wt[,4] < 0.01),]
head(sig_wt)
dim(sig_wt)
```

There are 2448 features that are significant different when comparing the two groups.

## Multivariate analysis, PCA

```r
if (!requireNamespace("pcaMethods", quietly = TRUE))
    BiocManager::install("pcaMethods")

suppressPackageStartupMessages(library("pcaMethods"))
```

```r
pc = pcaMethods::pca(t(norm_d1), nPcs = 9, method = "svd", scale = "pareto", center = TRUE)

pca = rbind.data.frame(labels_d1, norm_d1) %>%
        t(.) %>%
        as.data.frame(.) %>%
        rownames_to_column(.) %>%
        as_tibble(.) %>%
        select(rowname, Label) %>%
        column_to_rownames(.,var = "rowname") %>%
        merge(., pcaMethods::scores(pc), by=0)
```

### Score plot 

```r
ggplot(pca, aes(PC1, PC2,colour=Label)) + geom_point() + stat_ellipse() +
        xlab(paste("PC1", round(pc@R2[1] * 100, digits = 1), "% of the variance")) +
        ylab(paste("PC2", round(pc@R2[2] * 100, digits = 1), "% of the variance"))
```

As is shown in score plot, there are clear separation between these two groups, and the first PC accounts for ~74% of total variance of the dataset. The ellipses represent 95% confidence intervals. 

### Scree plot

```r
var_explained_pc = data.frame(PC = paste0("PC", 1:9), var_explained = (pc@sDev)^2/sum((pc@sDev)^2), eigenvalues = pc@sDev)
var_explained_pc %>%
        ggplot(aes(x=PC,y=eigenvalues, group=1))+
        geom_point(size=4)+
        geom_line()+
        labs(title="Scree plot: PCA on scaled data")
```

As PC1 accounts for majority of the variance in the dataset, we should take a look at what features count heavier in this PC. 

### Loading plot

```r
options(repr.plot.width=15, repr.plot.height=10)
barplot( pc@loadings[,1], main = "PC1 loading plot", las = 2, cex.names=0.6)
```

Don't panic! It looks messy because we have 2448 (significant) features here in this dataset. Contribution of each feature to PC1 is ranging from -0.04 to 0.04. We can try to visualize features that contribution heaviest to PC1, say > 0.035. 

```r
barplot( sort( pc@loadings[,1][which(abs(pc@loadings[,1]) > 0.035)], decreasing = FALSE), main = "PC1 loading plot", las = 2, cex.names=0.6)
```
